```python
def detect_text(path):
    """Detects text in the file."""
    from google.cloud import vision
    import io
    client = vision.ImageAnnotatorClient()

    with io.open(path, 'rb') as image_file:
        content = image_file.read()
        
    image = vision.Image(content=content)

    response = client.text_detection(image=image)
    texts = response.text_annotations
    
    result = ''
    for idx, text in enumerate(texts):
        if idx > 0: result += text.description 
    return result

    # if response.error.message:
    #     raise Exception(
    #         '{}\nFor more info on error messages, check: '
    #         'https://cloud.google.com/apis/design/errors'.format(
    #             response.error.message))
```

```python
# 파일이름을 오름차순으로 정렬하는 함수

def file_sorted(files):
    temp = [re.split(r"([0-9]+)", s) for s in files]
    sort = sorted(temp, key = lambda x: (x[0].lower(), int(x[1])))
    
    return [''.join(s) for s in sort]
```

```python
# 파일경로를 리스트로 만드는 함수

def make_file_paths(path):
    img_names = os.listdir(path)
    file_names = file_sorted(img_names)
    file_paths = []
    for file_name in file_names:
        file_paths.append(path + file_name)

    return file_paths
```

```python
def preprocess_sentence(sentence):
    import re
    # (a-z, A-Z, ㄱ-ㅎ, ㅏ-ㅣ, 가-힣)를 제외한 모든 문자를 공백인 ' '로 대체
    # sentence = re.sub(r"[^a-zA-Zㄱ-ㅣ가-힣0-9]+", " ", sentence)
    sentence = re.sub('[-=+,「#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','', sentence)
    sentence = sentence.replace(' ', '') # 공백 제거
    sentence = sentence.strip()
    return sentence
```

```python
for idx, sentence in enumerate(df1['책제목']):
    df1['책제목'][idx] = preprocess_sentence(sentence)
for idx, sentence in enumerate(df1['result']):
    df1['result'][idx] = preprocess_sentence(sentence)
for idx, sentence in enumerate(df1['result_resize']):
    df1['result_resize'][idx] = preprocess_sentence(sentence)
for idx, sentence in enumerate(df1['result_gray']):
    df1['result_gray'][idx] = preprocess_sentence(sentence)
for idx, sentence in enumerate(df1['result_resize2']):
    df1['result_resize2'][idx] = preprocess_sentence(sentence) 
```
