```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
epoch100 = pd.read_csv('results_epoch100.csv')
epoch150 = pd.read_csv('results_epoch150.csv')
epoch200 = pd.read_csv('results_epoch200.csv')
```

```python
epoch100.head()
```

```python
plot100 = epoch100.plot()
plt.show()
```

```python
#속성값 정렬
epoch100_val_loss = epoch150[['        val/obj_loss','        val/box_loss','        val/cls_loss']]
epoch100_train_loss = epoch150[['      train/box_loss','      train/obj_loss','      train/cls_loss']]
epoch100_metrics = epoch150[['   metrics/precision','      metrics/recall','metrics/mAP_0.5:0.95']]


```

```python
# Val loss plotting
plot_e100_val_loss = epoch100_val_loss.plot()
plt.title('Epoch 100 - val')

# Train loss plotting
plot_e100_train_loss = epoch100_train_loss.plot()
plt.title('Epoch 100 - train')

# Metrics plotting
plot_e100_metric = epoch100_metrics.plot()
plt.title('Epoch 100 - metrics')

plt.show()
```

```python
# calculating total loss
epoch100_val_loss['total_loss'] = epoch100_val_loss.sum(axis=1)
```
