```python
import os


label_path = '/content/drive/MyDrive/AIFFEL/OCR/Dataset/train_data_ver2/stage2/label/'

file_num = [i for i in range(401, 801)]
file_name = f'/image_{file_num[0]}.txt'

file_path = label_path + file_name

with open(file_path, 'r') as f:
  lines = f.readline().split()
  print('before: ', lines)
  lines[0] = '0'
  print('after: ', lines)

with open(label_path + 'label_test.txt', 'w') as f:
  f.write(' '.join(lines))
```

```python

for i in file_num:
  try:
    file_name = f'image_{i}.txt'
    file_path = label_path + file_name

    with open(file_path, 'r') as f:
      lines = f.readline().split()
      print('before: ', lines)
      lines[0] = '0'
      print('after: ', lines)

    with open(file_path, 'w') as f:
      print(file_name)
      f.write(' '.join(lines))
      
  except FileNotFoundError:
    pass
```
